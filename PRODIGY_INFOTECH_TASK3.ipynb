```python
# Data visualization
def add_value_labels(ax,spacing=10):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3
        
        va='bottom' if y>=0 else 'top'
        label= '()'.format(y) 
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

```python
# create a bar plot
plt.figure(figsize=(20,20))
plt.rcParams['axes.facecolor']=="blue"

plt.subplot(2,2,1)
plot=sns.barplot(x='sepal.length',y="variety",data=df,palette="viridis")
plt.title("IRIS")
plt.xlabel("SEPAL.LENGTH")
plt.ylabel("VARIETY")
add_value_labels(plot)

plt.subplot(2,2,2)
plot=sns.barplot(x='sepal.width',y="variety",data=df,palette="viridis")
plt.title("IRIS")
plt.xlabel("SEPAL.WIDTH")
plt.ylabel("VARIETY")
add_value_labels(plot)

plt.subplot(2,2,3)
plot=sns.barplot(x='petal.length',y="variety",data=df,palette="viridis")
plt.title("IRIS")
plt.xlabel("PETAL.LENGTH")
plt.ylabel("VARIETY")
add_value_labels(plot)

plt.subplot(2,2,4)
plot=sns.barplot(x='petal.width',y="variety",data=df,palette="viridis")
plt.title("IRIS")
plt.xlabel("PETAL.WIDTH")
plt.ylabel("VARIETY")
add_value_labels(plot)


plt.tight_layout()
```
